```python
import xgboost
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score
```

```python
X_train, X_test, y_train, y_test =  train_test_split(x, y ,test_size=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor()
parameters = {'learning_rate': [0.01,0.02, 0.05],
              'subsample'    : [0.2, 0.25],
              'n_estimators' : [800, 1000, 1100],
              'max_depth'    : [4,6,8] 
             }
grid_xgboost = GridSearchCV(estimator=model, param_grid = parameters, cv = 2, n_jobs=-1)
grid_xgboost.fit(X_train, y_train)
```

```python
grid_xgboost.best_params_
```

```python
predictions = grid_xgboost.predict(X_test)
predictions[predictions < 0] = 0
```

```python
print(explained_variance_score(predictions,y_test))
```

```python
[p for p in predictions if p<0]
```

```python
from sklearn.metrics import r2_score
print("R2 score: {}".format(r2_score(y_test, predictions)))
```

```python
boost_test = GradientBoostingRegressor()
boost_test.fit(X_train, y_train)

predictions = boost_test.predict(X_test)
print(explained_variance_score(predictions,y_test))
```
